```python
# Import all libraries.
import cloudscraper as cs
from bs4 import BeautifulSoup as beas
import pandas as pd
```

```python
# Create a scraper with a delay of 10 seconds and a custom user agent
scraper = cs.create_scraper(delay=10, browser={'custom':'ScraperBot/1.0'})

# Define the Amazon URL for laptops search
amazon_url = 'https://www.amazon.in/s?k=laptop&ref=nb_sb_ss_ts-doa-p_2_6'

# Send a request to the Amazon URL using the scraper
res = scraper.get(amazon_url)

print(res)
```

```python
# Creating a BeautifulSoup object to parse the HTML content.
soup  = beas(res.text,"html.parser")

# Find all <span> elements with the class "a-price-whole".
price = soup.findAll("span",{"class":"a-price-whole"})

# Initialize an empty list to store the extracted price values.
all_price = []

# Append stripped text from found <span> elements..
for i in range(len(price)):
 all_price.append(price[i].text.strip())
all_price
```

```python

soup = beas(res.text, "html.parser")
price = soup.findAll("span", {"class": "a-price-whole"})
all_price = []

# Convert price values to integers
for i in range(len(price)):
    price_value = price[i].text.strip().replace(',', '')  # Remove commas from the price
    all_price.append(int(price_value))

all_price
```

```python
# Creating a BeautifulSoup object to parse the HTML content.
soup  = beas(res.text,"html.parser")

# Find all <span> elements with the specified class for product names.
name = soup.findAll("span",{"class":"a-size-medium a-color-base a-text-normal"})

# Initialize an empty list to store all the product names.
all_name = []

# Append stripped text from found <span> elements..
for i in range(len(name)):
 all_name.append(name[i].text.strip())
all_name
```

```python
# Creating a dataframe.
df = pd.DataFrame({'Product_Name': all_name, 'Price': all_price})

# Export the data to csv file.
df.to_csv('Laptops.csv', index=False, encoding='utf-8')
df
```
